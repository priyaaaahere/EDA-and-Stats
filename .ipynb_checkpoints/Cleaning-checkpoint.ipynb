## Handling Duplicates

```python
# Handling Duplicate Rows
import pandas as pd

# Example DataFrame
data = {'A': [1, 2, 2, 4], 'B': [1, 2, 2, 4], 'C': [1, 2, 2, 4]}
df = pd.DataFrame(data)

# Original DataFrame
print("Original DataFrame:")
print(df)
```

```python
# Identifying duplicate rows
duplicate_rows = df[df.duplicated()]
print("\nDuplicate Rows:")
print(duplicate_rows)
```

```python
# Removing duplicate rows
df_no_duplicates = df.drop_duplicates() #deleting
print("\nDataFrame after removing duplicate rows:")
print(df_no_duplicates)
```

# Outliers

##### Identifying outliers before handling outliers, you need to identify them. Common methods include:
##### Z-score: Data points with Z-score greater than 3 or less than -3 are often considered outliers.
##### IQR (Interquartile Range): Data points that fall below lowe(Q1 - 1.5 x IQR Q1 - 1.5 x IQR) or above upper(Q3 + 1.5 x IQR Q3 + 1.5 x IQR) where IQR=Q3-Q1
##### Visualization: Use box plots, scatter plots. histograms to visually detect outliers


```python
import pandas as pd
import numpy as np
from scipy import stats

#Example Dataframe
data={'A':[1,2,3,100],'B':[2,3,4,5]}
df=pd.DataFrame(data)

#Original dataframe
print("Original Dataframe:")
print(df)

#Detecting outliers using z-score
z_scores=np.abs(stats.zscore(df))
print("\nZ-scores for each data point:")
print(z_scores)

#Removing outliers (where Z-score>3)
df_no_outliers=df[(z_scores<3).all(axis=1)]
print("\nDataframe after removing outliers:")
print(df_no_outliers)
```

```python
# Identify outliers using IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1


# Filtering out outliers
df_no_outliers = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]



# Original DataFrame
print("Original DataFrame:")
print(df)

# DataFrame after removing outliers
print("\nDataFrame after removing outliers:")
print(df_no_outliers)
```

# Handling Inconsistent Data Types


```python
#Handling Inconsistent Data Types
import pandas as pd

# Example DataFrame
data = {'A': ['1', '2', 'three', '4'], 'B': ['1.1', '2.2', 'three', '4.4']}
df = pd.DataFrame(data)

# Original DataFrame
print("Original DataFrame:")
print(df)
```

```python
# Converting to numeric, forcing invalid parsing to NaN
df['A'] = pd.to_numeric(df['A'], errors='coerce') #errors='coerce': Converts invalid parsing to NaN (Not a Number).
df['B'] = pd.to_numeric(df['B'], errors='coerce')
print("\nDataFrame after converting columns to numeric (forcing invalid to NaN):")
print(df)

```

```python
# Dropping or filling rows where conversion resulted in NaN
df_cleaned = df.dropna()
print("\nDataFrame after dropping rows with NaN values:")
print(df_cleaned)
```

# Handling Inconsistent Data Formats

```python
import pandas as pd
```

```python
#Example DataFrame
data={'Date':['01/01/2020','2020-02-01','March 3,2020']}
df=pd.DataFrame(data)
```

```python
#Original Dataframe
print("Original DataFrame: ")
print(df)
```

```python
#Converting to a consistent data format
df['Date']=pd.to_datetime(df['Date'],errors='coerce')
print("\nDataFrame after converting to a consistent date format:")
print(df)
```

```python
pd.to_datetime(df['Date'], format=("%m/%d/%Y", "%Y-%m-%d", "%B %d, %Y"), errors='coerce')
```

```python
df['Formatted_Date'] = df['Date'].dt.strftime('%B %d, %Y')
df
```

```python
df['Formatted_Date'] = df['Date'].dt.strftime('%d-%Y-%m')
df
```
