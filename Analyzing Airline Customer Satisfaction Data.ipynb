### Data Description

 0. Serial Number: A serial number is a unique identifier assigned to an individual item or product, typically used for tracking, authentication, and record-keeping purposes.
 
 1.   id         : Refers to a unique identifier for each record or entry in the dataset                  
 
 2.   Gender     : Gender refers to the social, cultural, and psychological traits and roles that a society considers                        
 3.   Customer Type    : The categorization of customers based on their loyalty to the airline                  
 
 4.   Age              : The numerical value representing the age of each customer              
 
 5. Type of Travel    : Indicates whether the travel is for personal or business purposes.
 
 6. Class   : Specifies the class of service (e.g., Economy, Business, First Class).
 
 7. Flight Distance   : The total distance covered during the flight, usually measured in miles or kilometers.
 
 8. Inflight wifi service   : Rating of the inflight Wi-Fi service provided to passengers.
 
 9. Departure/Arrival time convenient   : Passenger satisfaction with the convenience of departure and arrival times.
 
 10. Ease of Online booking   : Evaluation of the simplicity and efficiency of the online booking process.
 
 11. Gate location  : Passenger feedback on the proximity and accessibility of the gate location.
 
 12. Food and drink   : Satisfaction with the food and beverage options available during the flight.
 
 13. Online boarding   : Experience with the online boarding process, such as mobile check-in.
 
 14. Seat comfort   : Passenger rating of the comfort level of their seat.
 
 15. Inflight entertainment   : Satisfaction with the variety and quality of inflight entertainment options.
 
 16. On-board service   : Assessment of the service provided by the cabin crew during the flight.
 
 17. Leg room service   : Rating of the space and comfort provided by the legroom.
 
 18. Baggage handling   : Passenger feedback on the efficiency and care taken with luggage handling.
 
 19. Checkin service    : Evaluation of the ease and efficiency of the check-in process.
 
 20. Inflight service   : Overall rating of the services offered during the flight, including attention from staff.
 
 21. Cleanliness    : Passenger satisfaction with the cleanliness of the aircraft.
 
 22. Departure Delay in Minutes   : The number of minutes the departure was delayed, if any.
 
 23. Arrival Delay in Minutes   : The number of minutes the arrival was delayed, if any.
 
 24. Satisfaction   : Overall customer satisfaction with the airline experience, often rated on a scale.

```python
# Dropping rows with any missing values
df_cleaned_rows = df.dropna() #df.dropna(axis=1)-->columns      df.dropna(axis=0)-->rows 
print("\nDataFrame after dropping rows with any missing values:")
df_cleaned_rows
```

```python
# Filling missing values with the mean of the column
df_filled_mean = df.fillna(df.mean())
print("\nDataFrame after filling missing values with the mean of the column:")
df_filled_mean
```

```python
df.isnull().sum()
```

```python
#To avoid dropping columns due to missing values, we'll replace null values with mean/median/mode
df.fillna({
    # Numerical columns
    'Customer ID': df['Customer ID'].median(),
    'Flight Distance': df['Flight Distance'].median(),
    'Departure Delay in Minutes': df['Departure Delay in Minutes'].median(),
    'Arrival Delay in Minutes': df['Arrival Delay in Minutes'].median(),
    'Age': df['Age'].median(),
    
    # Categorical columns->mode() 
    'Gender': df['Gender'].mode()[0],
    'Customer Type': df['Customer Type'].mode()[0],
    'Type of Travel': df['Type of Travel'].mode()[0],
    'Class': df['Class'].mode()[0],
    'Inflight wifi service': df['Inflight wifi service'].mode()[0],
    'Departure/Arrival time convenient': df['Departure/Arrival time convenient'].mode()[0],
    'Ease of Online booking': df['Ease of Online booking'].mode()[0],
    'Gate location': df['Gate location'].mode()[0],
    'Food and drink': df['Food and drink'].mode()[0],
    'Online boarding': df['Online boarding'].mode()[0],
    'Seat comfort': df['Seat comfort'].mode()[0],
    'Inflight entertainment': df['Inflight entertainment'].mode()[0],
    'On-board service': df['On-board service'].mode()[0],
    'Leg room service': df['Leg room service'].mode()[0],
    'Baggage handling': df['Baggage handling'].mode()[0],
    'Checkin service': df['Checkin service'].mode()[0],
    'Inflight service': df['Inflight service'].mode()[0],
    'Cleanliness': df['Cleanliness'].mode()[0],
    'satisfaction': df['satisfaction'].mode()[0],
}, inplace=True)
```
